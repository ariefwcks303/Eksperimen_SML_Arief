```python
plt.figure(figsize=(10, 8))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Antar Fitur Numerik')
plt.show()
```

# **5. Data Preprocessing**

Pada tahap ini, data preprocessing adalah langkah penting untuk memastikan kualitas data sebelum digunakan dalam model machine learning.

Jika Anda menggunakan data teks, data mentah sering kali mengandung nilai kosong, duplikasi, atau rentang nilai yang tidak konsisten, yang dapat memengaruhi kinerja model. Oleh karena itu, proses ini bertujuan untuk membersihkan dan mempersiapkan data agar analisis berjalan optimal.

Berikut adalah tahapan-tahapan yang bisa dilakukan, tetapi **tidak terbatas** pada:
1. Menghapus atau Menangani Data Kosong (Missing Values)
2. Menghapus Data Duplikat
3. Normalisasi atau Standarisasi Fitur
4. Deteksi dan Penanganan Outlier
5. Encoding Data Kategorikal
6. Binning (Pengelompokan Data)

Cukup sesuaikan dengan karakteristik data yang kamu gunakan yah. Khususnya ketika kami menggunakan data tidak terstruktur.

```python
def clean_and_preprocess_dataframe(df_input):
    df_processed = df_input.copy()
    print(f"🎬 Memulai proses preprocessing pada DataFrame dengan {len(df_processed)} baris.")

    # 1. Handling Duplicates
    df_processed = df_processed.drop_duplicates().copy()
    print(f"   ↳ Data duplikat dihapus. Sisa baris: {len(df_processed)}")

    # 2. Encoding Target (placement_status) -> 1 jika Placed, 0 jika Not Placed
    if 'placement_status' in df_processed.columns:
        df_processed['target'] = df_processed['placement_status'].apply(lambda x: 1 if x == 'Placed' else 0)
        df_processed = df_processed.drop(columns=['placement_status'])
    else:
        raise KeyError("Kolom 'placement_status' sebagai target tidak ditemukan!")

    # 3. Handling Outliers (IQR Capping)
    numerical_cols = ['study_hours', 'attendance', 'sleep_hours', 'internet_usage',
                      'assignments_completed', 'previous_score', 'exam_score']

    print("   ↳ Melakukan IQR Capping pada fitur numerik...")
    for col in numerical_cols:
        Q1 = df_processed[col].quantile(0.25)
        Q3 = df_processed[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_processed[col] = df_processed[col].clip(lower=lower_bound, upper=upper_bound)

    # 4. Binning & Label Encoding (study_hours & previous_score)
    print("   ↳ Membuat fitur binning...")
    # study_hours -> Low, Medium, High -> 0, 1, 2
    study_hours_bins = pd.cut(df_processed['study_hours'], bins=3, labels=['Low', 'Medium', 'High'], include_lowest=True)
    df_processed['study_hours_encoded'] = LabelEncoder().fit_transform(study_hours_bins)

    # previous_score -> Q1, Q2, Q3, Q4 -> 0, 1, 2, 3
    prev_score_bins = pd.qcut(df_processed['previous_score'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
    df_processed['previous_score_encoded'] = LabelEncoder().fit_transform(prev_score_bins)

    # 5. Fitur Standardisasi (StandardScaler)
    print("   ↳ Melakukan Standardisasi dengan StandardScaler...")
    scaler = StandardScaler()
    df_processed[numerical_cols] = scaler.fit_transform(df_processed[numerical_cols])

    # 6. Menyusun Urutan Kolom Final
    # Menempatkan target di paling ujung kanan demi kerapian data modeling
    feature_cols = [col for col in df_processed.columns if col != 'target']
    final_df = df_processed[feature_cols + ['target']]

    print(f"✅ Preprocessing Selesai! Dimensi Data Final: {final_df.shape}")
    return final_df
```

```python
# Jalankan fungsi preprocessing pada DataFrame yang telah dimuat
df_cleaned = clean_and_preprocess_dataframe(df)

# Tampilkan beberapa baris pertama dari DataFrame yang sudah diproses
print("\nBerikut 5 baris pertama dari dataset yang telah diproses:")
display(df_cleaned.head())

# Simpan dataset yang sudah diproses
OUTPUT_PATH = 'student_dataset_clean.csv'
df_cleaned.to_csv(OUTPUT_PATH, index=False)
print(f"Dataset yang sudah dibersihkan dan diproses disimpan ke: {OUTPUT_PATH}")
```
